### Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('classic')
```

### Data

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(12, 7))
fig.tight_layout()

bplot1 = ax[0].boxplot(
    cars['horsepower'],
    patch_artist=True,
    boxprops=dict(facecolor='pink', color='red'),
    whiskerprops=dict(color='red'),
    capprops=dict(color='red'),
    flierprops=dict(marker='^', markerfacecolor='red', markersize=8, markeredgecolor='none')
)
ax[0].set_title('Horsepower')
pad = (cars['horsepower'].max() - cars['horsepower'].min()) * 0.05
ax[0].set_ylim(cars['horsepower'].min()-pad, cars['horsepower'].max()+pad)

ax[1].boxplot(
    cars['weight'],
    patch_artist=True,
    boxprops=dict(facecolor='lightblue', color='blue'),
    whiskerprops=dict(color='blue'),
    capprops=dict(color='blue'),
)
ax[1].set_title('Weight')
pad = (cars['weight'].max() - cars['weight'].min()) * 0.05
ax[1].set_ylim(cars['weight'].min()-pad, cars['weight'].max()+pad)

ax[2].boxplot(
    cars['acceleration'],
    patch_artist=True,
    boxprops=dict(facecolor='lightgreen', color='green'),
    whiskerprops=dict(color='green'),
    capprops=dict(color='green'),
    flierprops=dict(marker='.', markerfacecolor='green', markersize=8, markeredgecolor='none')
)
ax[2].set_title('Acceleration')
pad = (cars['acceleration'].max() - cars['acceleration'].min()) * 0.05
ax[2].set_ylim(cars['acceleration'].min()-pad, cars['acceleration'].max()+pad)

ax[3].boxplot(
    cars['model_year'],
    patch_artist=True,
    boxprops=dict(facecolor='#ffb999', color='orange'),
    whiskerprops=dict(color='orange'),
    capprops=dict(color='orange'),
)
ax[3].set_title('Model year')
pad = (cars['model_year'].max() - cars['model_year'].min()) * 0.05
ax[3].set_ylim(cars['model_year'].min()-pad, cars['model_year'].max()+pad)

plt.show()
```
