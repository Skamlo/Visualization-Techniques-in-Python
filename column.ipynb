### Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
plt.figure(figsize=(12, 3))
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot.bar(color='purple')
```

#### Kilka słupków w jedej kolumnie

```python
plt.figure(figsize=(12, 3))
data = pd.DataFrame(
    np.random.rand(6, 4),
    index = ['one', 'two', 'three', 'four', 'five', 'six'],
    columns = pd.Index(['A', 'B', 'C', 'D'], name='Rodzja')
)

data.plot.bar(color=['#ff0000', '#ff3333', '#ff6666', '#ff9999'], figsize=(12, 3))
data.plot.barh(color=['#ff0000', '#ff3333', '#ff6666', '#ff9999'], figsize=(12, 3), stacked=True)
```
