### Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('white')
```

### Data

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
sns.displot(
    data=cars,
    x='weight',
    bins=50,
    kde=True,
    height=4,
    aspect=3
)
plt.show()
```

```python
sns.displot(
    data=cars,
    x='weight',
    y='horsepower',
    kind='kde',
    hue='origin',
    col='origin',
    aspect=0.75,
    palette='crest',
    fill=True
)
plt.show()
```
