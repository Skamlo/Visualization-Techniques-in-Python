```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

```python
cars = sns.load_dataset('mpg').dropna().drop(['origin', 'name'], axis=1)
```

#### Correlation heat map

```python
shape = np.ones_like(cars.corr())
mask = np.array(np.triu(shape) - np.identity(len(shape)), dtype=np.bool_)

sns.heatmap(
    cars.corr(),
    cmap='inferno',
    annot=True,
    mask=mask
)
plt.show()
```
