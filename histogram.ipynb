### Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from PIL import Image

plt.style.use('ggplot')
```

### Data

```python
data1 = np.random.randn(1000)
data2 = np.random.randn(2500) * 1.5 + 4
```

### Besics

```python
bins = np.linspace(data1.min(), data2.max(), 100)

plt.figure(figsize=(12, 5))
plt.hist(data1, bins = bins, alpha = 0.85, color = 'brown')
plt.hist(data2, bins=bins, alpha = 0.85, color = 'darkgreen')
plt.show()
```

### Animated histogram

```python
for i in range(1, 151):
    ax = plt.hist(
        np.random.randn(int(10000 * (1 + i/150))) * (1 + i/150) + i/37.5,
        bins=100
    )
    plt.xlim(-4, 9)
    plt.ylim(0, 750)

    plt.savefig(f'hist imgs/img{i:003}', dpi=100, facecolor='white')
    plt.close()
```

```python
frames = [Image.open(f"hist imgs/{image}") for image in os.listdir('hist imgs')]
frame_one = frames[0]
frame_one.save(
    "histAnimation.gif",
    format="GIF",
    append_images=frames,
    save_all=True,
    duration=int(1000/30),
    loop=0
)
```
