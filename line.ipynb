### Libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

#### Podstawy

```python
data = np.cumsum(np.random.randn(250))

plt.figure(figsize=(12, 5))
plt.plot(
    data,
    linewidth = 1,
    linestyle = 'dashed',
    marker = "_",
    color = 'purple'
)
plt.title('Jakiś wykres XD')
plt.ylabel('Oś y')
plt.xlabel('Oś x')
plt.show()
```

#### Styl steps-post

```python
data = np.random.randn(30)
plt.figure(figsize=(12, 6))
plt.plot(data, color='lightblue', linestyle='--', label='Styl domyślny')
plt.plot(data, color='blue', linestyle='-', drawstyle='steps-post', label='Styl steps-post')
plt.legend(loc='best')
plt.show()
```

#### Obrazy na wykresie

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(1, 1, 1)

np.random.seed(42)
data = np.random.randn(1000).cumsum()

plt.plot(np.arange(1000), data, color='red', linewidth=1)

crisisData = [
    (427, "Górka"),
    (630, "Druga górka")
]

for pos, label in crisisData:
    ax.annotate(
        label,
        xy=(pos, data[pos] + 2),
        xytext=(pos, data[pos] + 8),
        arrowprops=dict(facecolor='red', headwidth=10, width=5, headlength=10)
    )

plt.show()
```
