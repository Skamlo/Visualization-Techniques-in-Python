### Libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

### Besic line graph

```python
data = np.cumsum(np.random.randn(250))

plt.figure(figsize=(12, 5))
plt.plot(
    data,
    linewidth = 1,
    linestyle = 'dashed',
    marker = "_",
    color = 'purple'
)
plt.title('Example line graph')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.show()
```

### steps-post style

```python
data = np.random.randn(30)
plt.figure(figsize=(12, 6))
plt.plot(data, color='lightblue', linestyle='--', label='default style')
plt.plot(data, color='blue', linestyle='-', drawstyle='steps-post', label='steps-post style')
plt.legend(loc='best')
plt.show()
```

### Pictures on graph

```python
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(1, 1, 1)

np.random.seed(42)
data = np.random.randn(1000).cumsum()

plt.plot(np.arange(1000), data, color='red', linewidth=1)

crisisData = [
    (427, "Local max"),
    (630, "Another local max")
]

for pos, label in crisisData:
    ax.annotate(
        label,
        xy=(pos, data[pos] + 2),
        xytext=(pos, data[pos] + 8),
        arrowprops=dict(facecolor='red', headwidth=10, width=5, headlength=10)
    )

plt.show()
```

### Area graph

```python
fn = lambda x: np.sin(x) * np.e**(-(x/7)**2)

x = np.arange(0, 10.05, 0.05)
y = fn(x)

plt.figure(figsize=(12, 6))
plt.plot(x, y, color='blue')
plt.fill_between(x, y, color='blue', alpha=0.25)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.cumsum(np.random.normal(1, 2, len(x)))
y2 = np.cumsum(np.random.normal(2, 2, len(x)))

plt.figure(figsize=(12, 6))
plt.plot(x, y1, color='#f52784')
plt.plot(x, y2, color='#36adf7')
plt.fill_between(x, y1, color='#f52784', alpha=0.25)
plt.fill_between(x, y2, color='#36adf7', alpha=0.25)
plt.show()
```
