### Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

### Data

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
plt.figure(figsize=(8, 6))

cylinders = [3, 4, 5, 6, 8]
colors = ['#b3ff00', '#ffd900', '#ff9900', '#ff4d00', '#ff0000']
markers = ['^', 's', 'p', 'h', '8']

for n, color, marker in zip(cylinders, colors, markers):
    cyl = cars[cars['cylinders'] == n]
    plt.scatter(cyl['weight'], cyl['horsepower'], color=color, marker=marker, s=20, label=f'{n} cyl.')

plt.title('Zestawienie wagi z ilością koni mechanicznych w samochodach')
plt.xlabel('Weight')
plt.ylabel('Horsepower')
plt.legend()
plt.show()
```
