### Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('classic')
```

### Data

```python
cars = sns.load_dataset('mpg').dropna()
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(12, 5))
fig.tight_layout()


violin1 = ax[0].violinplot(
    cars['horsepower'],
    showmedians=True
)
for pc in violin1['bodies']:
    pc.set_facecolor('#400080')
ax[0].set_title('Horsepower')
pad = (cars['horsepower'].max() - cars['horsepower'].min()) * 0.05
ax[0].set_ylim(cars['horsepower'].min()-pad, cars['horsepower'].max()+pad)

violin2 = ax[1].violinplot(
    cars['weight'],
    showmedians=True
)
for pc in violin2['bodies']:
    pc.set_facecolor('#7000e0')
ax[1].set_title('Weight')
pad = (cars['weight'].max() - cars['weight'].min()) * 0.05
ax[1].set_ylim(cars['weight'].min()-pad, cars['weight'].max()+pad)

violin3 = ax[2].violinplot(
    cars['acceleration'],
    showmedians=True
)
for pc in violin3['bodies']:
    pc.set_facecolor('#962eff')
ax[2].set_title('Acceleration')
pad = (cars['acceleration'].max() - cars['acceleration'].min()) * 0.05
ax[2].set_ylim(cars['acceleration'].min()-pad, cars['acceleration'].max()+pad)

violin4 = ax[3].violinplot(
    cars['model_year'],
    showmedians=True
)
for pc in violin4['bodies']:
    pc.set_facecolor('#b770ff')
ax[3].set_title('Model year')
pad = (cars['model_year'].max() - cars['model_year'].min()) * 0.05
ax[3].set_ylim(cars['model_year'].min()-pad, cars['model_year'].max()+pad)


plt.show()
```
